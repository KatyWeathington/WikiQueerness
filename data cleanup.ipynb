```python
import pandas as pd
```

```python
#wikidataSupportDataOccupation.csv
df=pd.read_csv('wikidataQueers21822.csv',keep_default_na=False,index_col=0)
df.head()
```

```python
len(df)
```

```python
df=df.drop_duplicates()
print(len(df))
df.head()
```

```python
qids=df['QID'].unique()
#print(qids)
```

```python
len(qids)
```

```python
dd=df.iloc[0:0]
dd.head()
```

```python
# dd.loc[q]=[d['Name'].unique(),
#            d['P21'].unique(),
#            d['P21ref'].unique(),
#           d['P91'].unique(),
#           d['P91ref'].unique(),
#           d['P6553'].unique(),
#           d['P6553ref'].unique(),
#           d['pronounsEn'].unique()]
```

```python
dd=df.iloc[0:0]
for q in qids:
    if len(df[df['QID']==q])==1:
        pass
    else:
        d=df[df['QID']==q]
        n=d['Name'].unique()
        if len(n)==0:
            n=''
        elif len(n)==1:
            n=n[0]
        
        p21=d['P21'].unique()
        if len(p21)==0:
            p21=''
        elif len(p21)==1:
            p21=p21[0]
            
        p21r=d['P21ref'].unique()
        if len(p21r)==0:
            p21r=''
        elif len(p21r)==1:
            p21r=p21r[0]
        
        p91=d['P91'].unique()
        if len(p91)==0:
            p91=''
        elif len(p91)==1:
            p91=p91[0]
        
        p91r=d['P91ref'].unique()
        if len(p91r)==0:
            p91r=''
        elif len(p91r)==1:
            p91r=p91r[0]
        
        p6553=d['P6553'].unique()
        if len(p6553)==0:
            p6553=''
        elif len(p6553)==1:
            p6553=p6553[0]
        
        p6553r=d['P6553ref'].unique()
        if len(p6553r)==0:
            p6553r=''
        elif len(p6553r)==1:
            p6553r=p6553r[0]
        
        pronouns=d['pronounsEn'].unique()
        if len(pronouns)==0:
            pronouns=''
        elif len(pronouns)==1:
            pronouns=pronouns[0]
            
        #break
        dd.loc[len(dd)]=[q,n,p21,p21r,p91,p91r,p6553,p6553r,pronouns]
        
        #dd.loc[len(dd)]=[q,d['Name'].unique(),d['P21'].unique(),d['P21ref'].unique(),d['P91'].unique(),d['P91ref'].unique(),d['P6553'].unique(),d['P6553ref'].unique(),d['pronounsEn'].unique()]
    #print(len(d))
    
```

```python
len(dd)
```

```python
dd.head()
```

```python
dMerged=pd.concat([df,dd])
```

```python
dz=dMerged.drop_duplicates(subset=['QID'],keep='last')
```
